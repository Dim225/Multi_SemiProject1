```python
plt.plot([1,2], [0.8,0.8], color = 'c', linewidth = 10)
plt.plot([1,2], [0.9,0.9], color = 'skyblue', linewidth = 10)
plt.plot([1,2], [1,1], color = 'aqua', linewidth = 10)
plt.plot([1,2], [1.1,1.1], color = 'lightseagreen', linewidth = 10)
plt.plot([1,2], [1.2,1.2], color = 'darkcyan', linewidth = 10)

plt.show()
```

```python
df.plot(x='예정일시', y='대기시간', kind = 'line', figsize = (25, 6),ylim = (0, 550), color = 'skyblue')
plt.xlabel('년도', fontsize =17)
plt.ylabel('대기시간', fontsize =15)
plt.title('2019-2022 대기시간 시계열 그래프', fontsize =25)
plt.grid(True, linestyle=':')
plt.savefig('2019-2022 대기시간 시계열 그래프', bbox_inches='tight' )
plt.show()
```
